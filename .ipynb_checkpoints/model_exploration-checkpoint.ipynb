### Scale data

```python
from sklearn.preprocessing import StandardScaler
absenteeism_scaler = StandardScaler()
```

```python
# import the libraries needed to create the Custom Scaler


from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

# create the Custom Scaler class

class CustomScaler(BaseEstimator,TransformerMixin): 
    
    # init 
    
    def __init__(self,columns,copy=True,with_mean=True,with_std=True):
        
        # scaler is nothing but a Standard Scaler object
#         self.scaler = StandardScaler(copy,with_mean,with_std)
        self.scaler = StandardScaler()
        # with some columns 'twist'
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        
        self.with_mean = with_mean
        self.with_std = with_std
        self.copy = copy

    # the fit method, which, again based on StandardScaler
    
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    # the transform method which does the actual scaling

    def transform(self, X, y=None, copy=None):
        
        # record the initial order of the columns
        init_col_order = X.columns
        
        # scale all features that you chose when creating the instance of the class
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        
        # declare a variable containing all information that was not scaled
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        
        # return a data frame which contains all scaled features and all 'not scaled' features
        # use the original order 
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

### Model 1: Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
log_param = {"C":np.logspace(-5,5,10), "penalty":["l1","l2"]}
clf_log = GridSearchCV(LogisticRegression(random_state=44,solver='liblinear'), log_param, cv=5)
clf_log.fit(x_train, y_train)
print(clf_log.best_params_) 
print(clf_log.best_estimator_.score(x_test,y_test))
```

```python
kfold = StratifiedKFold(n_splits=5,shuffle = True,random_state=440)


# x_train_copy = x_train.copy()

# n will be the maximum max_depth I consider
n = 20

accs = np.zeros((5,n))
precs = np.zeros((5,n))
recs = np.zeros((5,n))

for j in range(1,n+1):
    clf_tree_1 = DecisionTreeClassifier(max_depth=j, random_state = 44,  criterion = 'entropy')
    
    i = 0
    for train_index,test_index in kfold.split(x_train,y_train):
        x_train_train, x_train_test = x_train.iloc[train_index],x_train.iloc[test_index]
        y_train_train, y_train_test = y_train[train_index],y_train[test_index]

        clf_tree_clone = clone(clf_tree_1)
        clf_tree_clone.fit(x_train_train,y_train_train)
        
        y_predict = clf_tree_clone.predict(x_train_test)
        accs[i,j-1] = accuracy_score(y_train_test,y_predict)
        precs[i,j-1] = precision_score(y_train_test,y_predict)
        recs[i,j-1] = recall_score(y_train_test,y_predict)
        i = i + 1
```

```python
plt.figure(figsize=(14,10))

plt.plot(range(1,n+1), np.mean(accs,axis=0),'b-o', label="Accuracy")
plt.plot(range(1,n+1), np.mean(precs,axis=0),'r-o', label="Precision")
plt.plot(range(1,n+1), np.mean(recs,axis=0),'g-o', label="Recall")

plt.ylabel("Mean CV Performance Score (Entropy)",fontsize=16)
plt.xlabel("Max Depth",fontsize=16)

plt.axvline(x=7,c='r')

plt.legend(fontsize=16)

plt.show()
```

```python
tf.random.set_seed(4)

model = Sequential()
model.add(Dense(16, input_dim=x_train.shape[1], activation='relu'))
# model.add(Dense(16, activation='softmax'))
# model.add(Dense(8, activation='relu'))
model.add(Dense(2, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(patience=2)

history = model.fit(x_train_train, 
                    y_train_train, 
                    epochs=200, 
                    callbacks = [early_stopping],
                    validation_data = (x_train_val,y_train_val), 
                    batch_size=40,
                    verbose=2
                   )
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
